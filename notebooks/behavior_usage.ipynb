```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from tkinter import *
from tkinter import filedialog
```

### labels: {0: pause, 1: rear, 2: groom, 3: left, 4: right, 5: straight, 6:other}

```python
label_names = ['pause', 'investigate 1', 'investigate 2', 'investigate 3', 'rear', 'straight']
```

## wt & ko basline

```python
# Create Tk root
root = Tk()
# Hide the main window
root.withdraw()
root.call('wm', 'attributes', '.', '-topmost', True)
files_baseline = filedialog.askopenfilename(multiple=True, title='load baseline labels csv')

%gui tk
```

```python
animal_id_baseline = [files_baseline[j].rpartition('/')[2].rpartition('_labels')[0] for j in range(len(files_baseline))]
```

```python
baseline_binarys = []
baseline_labels = []
for baseline_file in files_baseline:
    # load each file
    baseline_binary = pd.read_csv(baseline_file, low_memory=False)
    baseline_binarys.append(baseline_binary)
    baseline_labels.append(np.argmax(np.array(baseline_binary.iloc[:, 1:]), axis=1))
```

```python
baseline_data_dict = {'animal_id': np.hstack([np.repeat(animal_id_baseline[i], len(baseline_labels[i]))
                           for i in range(len(baseline_labels))]),
             'behavior': np.hstack(baseline_labels)}
baseline_labels_df = pd.DataFrame(data=baseline_data_dict)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
ax = sns.histplot(data=baseline_labels_df,
                  x='behavior',
                  hue='animal_id',
                  multiple="dodge",
                  stat = 'percent', common_norm=False)
ax.set_xticks(np.arange(0, len(label_names)))
ax.set_xticklabels(label_names)
ax.set_ylim([0, 20])
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 3))
ax = sns.histplot(data=baseline_labels_df,
                  x='behavior',
                  hue='animal_id',
                  multiple="dodge",
                  stat = 'percent', common_norm=False)
ax.set_xticks(np.arange(0, len(label_names)))
ax.set_xticklabels(label_names)
ax.set_ylim([0, 20])
plt.show()
```
